```python
# Exercise 1: 3D Plotting and Pie Chart Creation

import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the DataFrame
df = pd.DataFrame({
    'Temperature': [20, 22, 24, 26, 28],
    'Rainfall': [1, 2, 1, 0, 1],
    'Crop Yield': [100, 120, 140, 160, 180]
})

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Temperature'], df['Rainfall'], df['Crop Yield'])

# Set labels
ax.set_xlabel('Temperature')
ax.set_ylabel('Rainfall')
ax.set_zlabel('Crop Yield')
ax.set_title('3D Scatter Plot of Temperature, Rainfall, and Crop Yield')

plt.show()
```

```python
# Part 2: Pie Chart

import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Get the counts of each embarkation point
embark_counts = titanic['embark_town'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(embark_counts, labels=embark_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Passengers Embarkation Points')
plt.show()
```
