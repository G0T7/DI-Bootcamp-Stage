```python
# Exercise 1: Install Matplotlib and Import
# Install matplotlib if not already installed
# !pip install matplotlib

# Import matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
# Exercise 2: Basic Line Plot

# Data for the line plot
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o')
plt.title("Product Demand vs Price")
plt.xlabel("Price")
plt.ylabel("Demand")
plt.grid(True)
plt.show()
```

```python
# Exercise 3: Basic Scatter Plot

# Data for the scatter plot
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', marker='x')
plt.title("Children’s Height vs Age")
plt.xlabel("Age (years)")
plt.ylabel("Height (cm)")
plt.grid(True)
plt.show()
```

```python
# Exercise 4: Modify Line Plot with Title and Labels

# Data for the line plot
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

# Create a line plot with title and labels
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', color='tab:blue')
plt.title("Product Demand vs Price")
plt.xlabel("Price")
plt.ylabel("Demand")
plt.grid(True)
plt.show()
```

```python
# Exercise 5: Modify Scatter Plot with Title and Labels

# Data for the scatter plot
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

# Create a scatter plot with title and labels
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', marker='x')
plt.title("Children’s Height vs Age")
plt.xlabel("Age (years)")
plt.ylabel("Height (cm)")
plt.grid(True)
plt.show()
```

```python
# Exercise 6: Save Plots as PNG Files

# Save the line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', color='tab:blue')
plt.title("Product Demand vs Price")
plt.xlabel("Price")
plt.ylabel("Demand")
plt.grid(True)
plt.savefig('line_plot.png')

# Save the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', marker='x')
plt.title("Children’s Height vs Age")
plt.xlabel("Age (years)")
plt.ylabel("Height (cm)")
plt.grid(True)
plt.savefig('scatter_plot.png')
```

```python
# Exercise 7: Plot Mathematical Functions

import numpy as np

# Data for mathematical functions
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# Line plot for sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, color='tab:blue', label='sin(x)')
plt.title("Mathematical Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot for cos(x)
plt.figure(figsize=(8, 6))
plt.scatter(x, y2, color='orange', marker='o', label='cos(x)')
plt.title("Mathematical Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np


# Data for mathematical functions
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 12))

# Line plot
ax1.plot(x, y1, color='tab:blue', label='sin(x)')
ax1.set_title('Line Plot: sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()

# Scatter plot
ax2.scatter(x, y2, color='orange', marker='o', label='cos(x)')
ax2.set_title('Scatter Plot: cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)
ax2.legend()

# Show plots
plt.tight_layout()
plt.show()
```

```python
# Exercise 9: Add Legends to Subplots

import numpy as np


# Data for mathematical functions
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 12))

# Line plot
ax1.plot(x, y1, color='tab:blue', label='sin(x)')
ax1.set_title('Line Plot with Legend')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()  # Adding legend here

# Scatter plot
ax2.scatter(x, y2, color='orange', marker='o', label='cos(x)')
ax2.set_title('Scatter Plot with Legend')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)
ax2.legend()  # Adding legend here

# Show plots
plt.tight_layout()
plt.show()

```
