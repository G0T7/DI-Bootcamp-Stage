```python
# Task 1: Plotting x vs y

# Step 1: Import the necessary packages
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Create an array x of 100 numbers ranging from 0 to 20 (inclusive)
x = np.linspace(0, 20, 100)

# Step 3: Create an array y which is the square of x
y = x ** 2

# Step 4: Plot x vs y
plt.plot(x, y)

# Step 5: Add title and labels, adjust limits
plt.title('Plot of x vs y')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 20)
plt.ylim(0, 400)

# Step 6: Save the plot as a .png file
plt.savefig('plot_x_vs_y.png')
plt.show()
```

```python
# Task 2: Scatter Plot of Item Price vs Quantity Ordered

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the dataset using pandas
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')

# Step 2: Clean the item_price column and transform it into a float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 3: Create a scatter plot presenting the relationship between the price of items and the quantity ordered
plt.scatter(chipo['item_price'], chipo['quantity'])

# Step 4: Add title and labels
plt.title('Scatter Plot of Item Price vs Quantity Ordered')
plt.xlabel('Item Price ($)')
plt.ylabel('Quantity Ordered')

# Step 5: Save the plot as a .png file
plt.savefig('scatter_item_price_vs_quantity.png')
plt.show()
```

```python
# Task 3: Bar Plot of Total Quantities for Top 5 Items

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the dataset using pandas
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')

# Step 2: Clean the item_price column and transform it into a float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 3: Group the ‘chipo’ data by item_name and calculate the total quantity for each item
item_totals = chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)

# Step 4: Create a bar plot that presents the total quantities for the top 5 items
item_totals.plot(kind='bar', color='skyblue')

# Step 5: Add title and labels
plt.title('Total Quantities for Top 5 Items')
plt.xlabel('Item Name')
plt.ylabel('Total Quantity')

# Step 6: Save the plot as a .png file
plt.savefig('bar_top_5_items.png')
plt.show()
```
