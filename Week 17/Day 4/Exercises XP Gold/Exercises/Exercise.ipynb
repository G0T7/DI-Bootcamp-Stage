```python
# Exercise 1: Line Graph from DataFrame

# Plotting a Line Graph
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
df = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temperature': [22, 24, 23, 24, 25, 26, 25],
    'Rainfall': [1, 0, 2, 1, 0, 0, 1]
})

# Set 'Day' as the index
df.set_index('Day', inplace=True)

# Plot the data
df.plot(kind='line', figsize=(10, 6))

# Add title and labels
plt.title('Daily Weather Data')
plt.xlabel('Day')
plt.ylabel('Value')
plt.grid(True)

# Show plot
plt.show()
```

```python
# Adding Error Bars

# Standard deviation
error = 1

# Plot the data with error bars
df.plot(kind='line', yerr=error, figsize=(10, 6))

# Add title and labels
plt.title('Daily Weather Data with Error Bars')
plt.xlabel('Day')
plt.ylabel('Value')
plt.grid(True)

# Show plot
plt.show()
```

```python
# Creating a Plot with Twin Axes

import numpy as np
import matplotlib.pyplot as plt

# Create some example data for twin axes
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot y1 on the primary y-axis
ax1.plot(x, y1, 'tab:blue', label='sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a twin Axes sharing the x-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, 'tab:orange', label='cos(x)')
ax2.set_ylabel('cos(x)', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Add title and legend
plt.title('Plot with Twin Axes')
fig.tight_layout()
plt.show()

```

```python
# Bar Graph with tab: Colors

# Bar graph using tab colors
df.plot(kind='bar', figsize=(10, 6), color=['tab:blue', 'tab:orange'])

# Add title and labels
plt.title('Daily Weather Data (Bar Graph)')
plt.xlabel('Day')
plt.ylabel('Value')
plt.grid(True)

# Show plot
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Example data
data = {
    'age': [22, 34, 28, 45, 32],
    'fare': [7.25, 71.83, 8.05, 53.1, 8.43],
    'survived': [0, 1, 1, 1, 0]
}
df = pd.DataFrame(data)

fig = px.scatter(df, x='age', y='fare', color='survived',
                 title="Age vs Fare",
                 labels={'age': 'Age', 'fare': 'Fare'},
                 color_continuous_scale=px.colors.sequential.Viridis)

fig.write_html("scatter_plot.html")
```
