```python
#Super Stack 

class Superstack:
    def __init__(self):
        self.stack = []
        self.increments = []
    
    def push(self, value: int):
        self.stack.append(value)
        self.increments.append(0)  # Initialize increment for new element
        self.print_top()
    
    def pop(self):
        if not self.stack:
            print("Empty")
            return
        
        if len(self.stack) > 1:
            self.increments[-2] += self.increments[-1]
        
        self.stack.pop()
        self.increments.pop()
        self.print_top()
    
    def inc(self, i: int, value: int):
        if self.stack:
            self.increments[i - 1] += value  # Increment the i-th position
        self.print_top()
    
    def print_top(self):
        if self.stack:
            print(self.stack[-1] + self.increments[-1])
        else:
            print("Empty")

# Example usage
operations = ['push 4', 'push 5', 'inc 2 1', 'pop', 'pop']
stack = Superstack()

for operation in operations:
    parts = operation.split()
    if parts[0] == 'push':
        stack.push(int(parts[1]))
    elif parts[0] == 'pop':
        stack.pop()
    elif parts[0] == 'inc':
        stack.inc(int(parts[1]), int(parts[2]))




#Big O Considerations
#Time Complexity:

#push: 𝑂(1)
#pop: 𝑂(1)
#inc: 𝑂(1)
#The print_top operation is 𝑂(1) and is called after every command.

#Space Complexity:
#The space complexity is 𝑂(𝑛) where 𝑛 is the number of operations, due to storing the stack and the auxiliary increments array.

#Scalability:
#Feasibility for Large Lists: The solution is efficient and feasible even for very large lists of operations (100K or 10M items) due to its constant time complexity for each operation.
#Most Intensive Operation: Without optimization, the inc operation would be the most intensive, but using the auxiliary increments array, we reduce its complexity to 𝑂(1)
```
