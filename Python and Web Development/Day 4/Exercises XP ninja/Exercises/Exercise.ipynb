```python
# Exercise 1: Formula

import math

# Fixed values
C = 50
H = 30

# Ask user for a comma-separated string of numbers
input_numbers = input("Enter comma-separated numbers: ")

# Split the input string into a list of numbers
numbers = [int(x) for x in input_numbers.split(',')]

# Calculate and print the result for each number
results = []
for D in numbers:
    Q = math.sqrt((2 * C * D) / H)
    results.append(int(Q))  # Convert to integer as in the example output

print(",".join(map(str, results)))
```

```python
# Exercise 2: List of Integers

# Define the list of numbers
numbers = [3, 47, 99, -80, 22, 97, 54, -23, 5, 7]  # Example list

# Print the list of numbers
print("List of numbers:", numbers)

# Print the list sorted in descending order
print("Sorted in descending order:", sorted(numbers, reverse=True))

# Print the sum of all the numbers
print("Sum of all numbers:", sum(numbers))

# Print a list containing the first and last numbers
print("First and last numbers:", [numbers[0], numbers[-1]])

# Print a list of numbers greater than 50
print("Numbers greater than 50:", [num for num in numbers if num > 50])

# Print a list of numbers smaller than 10
print("Numbers smaller than 10:", [num for num in numbers if num < 10])

# Print a list of squared numbers
print("Squared numbers:", [num ** 2 for num in numbers])

# Print numbers without duplicates
unique_numbers = list(set(numbers))
print("Numbers without duplicates:", unique_numbers)
print("Number of unique numbers:", len(unique_numbers))

# Print the average of all the numbers
print("Average of numbers:", sum(numbers) / len(numbers))

# Print the largest number
print("Largest number:", max(numbers))

# Print the smallest number
print("Smallest number:", min(numbers))

# Bonus: Sum, average, largest, and smallest without built-in functions
total = 0
count = 0
largest = numbers[0]
smallest = numbers[0]

for num in numbers:
    total += num
    count += 1
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

average = total / count
print("Sum (without built-in functions):", total)
print("Average (without built-in functions):", average)
print("Largest number (without built-in functions):", largest)
print("Smallest number (without built-in functions):", smallest)

# Bonus: Generate 10 random integers
import random

random_numbers = [random.randint(-100, 100) for _ in range(10)]
print("Random numbers:", random_numbers)

# Bonus: Random number of integers
num_integers = random.randint(50, 100)
random_numbers = [random.randint(-100, 100) for _ in range(num_integers)]
print(f"Random number of integers ({num_integers}):", random_numbers)
```

```python
# Exercise 3: Working on a Paragraph

# Example paragraph
paragraph = """
Python is an interpreted high-level general-purpose programming language. Its design philosophy emphasizes code readability with the use of significant indentation. Python is dynamically typed and garbage-collected. It supports multiple programming paradigms, including structured (particularly procedural), object-oriented and functional programming. Python is often described as a "batteries included" language due to its comprehensive standard library.
"""

# Count characters, sentences, words, and unique words
import string
from collections import Counter

# Remove punctuation for accurate word count
translator = str.maketrans('', '', string.punctuation)
clean_paragraph = paragraph.translate(translator)

# Count characters
num_characters = len(paragraph.replace(" ", ""))

# Count sentences
num_sentences = paragraph.count('.') + paragraph.count('!') + paragraph.count('?')

# Count words
words = clean_paragraph.split()
num_words = len(words)

# Count unique words
unique_words = set(words)
num_unique_words = len(unique_words)

# Print results
print(f"Number of characters: {len(paragraph)}")
print(f"Number of sentences: {num_sentences}")
print(f"Number of words: {num_words}")
print(f"Number of unique words: {num_unique_words}")

# Bonus: Number of non-whitespace characters
num_non_whitespace = len(paragraph.replace(" ", ""))
print(f"Number of non-whitespace characters: {num_non_whitespace}")

# Bonus: Average number of words per sentence
average_words_per_sentence = num_words / num_sentences if num_sentences > 0 else 0
print(f"Average number of words per sentence: {average_words_per_sentence:.2f}")

# Bonus: Number of non-unique words
word_counts = Counter(words)
num_non_unique_words = sum(count - 1 for count in word_counts.values() if count > 1)
print(f"Number of non-unique words: {num_non_unique_words}")
```

```python
# Exercise 4: Frequency Of The Words

from collections import Counter
import re

# Example input
text = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."

# Normalize and split the text into words
words = re.findall(r'\b\w+\b', text)

# Count the frequency of each word
word_counts = Counter(words)

# Print the frequencies
for word, count in word_counts.items():
    print(f"{word}:{count}")
```
