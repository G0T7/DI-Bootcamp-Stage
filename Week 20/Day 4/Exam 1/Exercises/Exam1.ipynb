```python
# Lists and Loops

# Using a list comprehension, generate a list of the squares of numbers from 1 to 10, but only include squares of even numbers.
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]

# Using the range function, create a list of numbers from 1 to 10, then print numbers that are divisible by both 2 and 3.
divisible_by_2_and_3 = [x for x in range(1, 11) if x % 2 == 0 and x % 3 == 0]
```

```python
#  Function Behavior with *args and **kwargs

def combine_words(*args, **kwargs):
    sentence = " ".join(args)
    if kwargs:
        sentence += ". " + " ".join(f"{value}" for key, value in sorted(kwargs.items()))
    return sentence

# Example:
print(combine_words("Hello", "world", second="is", third="great!", first="Python"))
# Expected Output: "Hello world. Python is great!"
```

```python
# Object-Oriented Programming (OOP)

class Vehicle:
    def __init__(self, type, brand, year):
        if not type or not brand or not year:
            raise ValueError("All attributes must be provided")
        self.type = type
        self.brand = brand
        self.year = year

    def __str__(self):
        return f"Vehicle Type: {self.type}, Brand: {self.brand}, Year: {self.year}"
```

```python
# OOP Inheritance and Decorators

class Car:
    def __init__(self, brand, model, mileage):
        self.brand = brand
        self.model = model
        self.mileage = mileage

    def car_details(self):
        return f"Brand: {self.brand}, Model: {self.model}, Mileage: {self.mileage}"

class ElectricCar(Car):
    def __init__(self, brand, model, mileage, battery_capacity):
        super().__init__(brand, model, mileage)
        self.__battery_capacity = battery_capacity

    @property
    def battery_capacity(self):
        return self.__battery_capacity

    @battery_capacity.setter
    def battery_capacity(self, value):
        if value > 0:
            self.__battery_capacity = value

    def car_details(self):
        return f"{super().car_details()}, Battery Capacity: {self.__battery_capacity} kWh"

class BankAccount:
    account_count = 0

    def __init__(self, account_holder, balance=0.0):
        self._account_holder = account_holder
        self._balance = balance
        BankAccount.account_count += 1

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount

    def view_balance(self):
        return self._balance

    @classmethod
    def get_account_count(cls):
        return cls.account_count

    @staticmethod
    def bank_policy():
        return "Bank policy: Maintain a minimum balance of $100."
```

```python
# Data Science

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a numpy array of shape (3,3) filled with integers from 1 to 9 using arange()
array = np.arange(1, 10).reshape(3, 3)

# Provided pandas DataFrame df:
df = pd.DataFrame({'A': [1, 'apple', 3, 4, 'banana'], 'B': [5, 6, 7, 8, 9]})

# Replace non-numeric values in column “A” with the mean of numeric values
numeric_mean = df['A'][df['A'].apply(lambda x: isinstance(x, (int, float)))].mean()
df['A'] = df['A'].apply(lambda x: numeric_mean if not isinstance(x, (int, float)) else x)

# Plot a histogram of the “A” column using matplotlib
df['A'].hist()
plt.title('Histogram of A')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Plot “A” and “B” columns of df using matplotlib. Add x-axis, y-axis labels, and a title.
plt.plot(df['A'], df['B'], marker='o')
plt.title('Plot of A vs B')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```
