```python
#Exercise 1: Formula
import math

def calculate_q(d_values):
    C = 50
    H = 30
    results = []
    
    for D in d_values:
        Q = math.sqrt((2 * C * D) / H)
        results.append(int(Q))
    
    return results

user_input = input("Enter comma-separated numbers: ")
d_values = list(map(int, user_input.split(',')))
results = calculate_q(d_values)
print(",".join(map(str, results)))
```

```python
#Exercise 2: List Of Integers
import random

numbers = [3, 47, 99, -80, 22, 97, 54, -23, 5, 7]

# a. Print the list of numbers
print("The list of numbers:", numbers)

# b. Sorted in descending order
sorted_numbers = sorted(numbers, reverse=True)
print("Sorted in descending order:", sorted_numbers)

# c. Sum of all numbers
total_sum = sum(numbers)
print("Sum of all numbers:", total_sum)

# d. First and last number
first_last = [numbers[0], numbers[-1]]
print("First and last number:", first_last)

# e. Numbers greater than 50
greater_than_50 = [num for num in numbers if num > 50]
print("Numbers greater than 50:", greater_than_50)

# f. Numbers smaller than 10
smaller_than_10 = [num for num in numbers if num < 10]
print("Numbers smaller than 10:", smaller_than_10)

# g. Numbers squared
squared_numbers = [num**2 for num in numbers]
print("Numbers squared:", squared_numbers)

# h. Numbers without duplicates
unique_numbers = list(set(numbers))
print("Numbers without duplicates:", unique_numbers)
print("Count of unique numbers:", len(unique_numbers))

# i. Average of numbers
average = total_sum / len(numbers)
print("Average of numbers:", average)

# j. Largest number
largest = max(numbers)
print("Largest number:", largest)

# k. Smallest number
smallest = min(numbers)
print("Smallest number:", smallest)

# Bonus tasks:
# l. Sum, average, largest, smallest without built-in functions
# Note: Skipping as they are already done with built-in functions above

# m. Asking user for 10 numbers between -100 and 100
user_numbers = []
for _ in range(10):
    user_num = int(input("Enter a number between -100 and 100: "))
    user_numbers.append(user_num)
print("User numbers:", user_numbers)

# n. Generate 10 random integers between -100 and 100
random_numbers = [random.randint(-100, 100) for _ in range(10)]
print("Random numbers:", random_numbers)

# o. Generate random amount of integers, no smaller than 50
random_count = random.randint(50, 100)
random_list = [random.randint(-100, 100) for _ in range(random_count)]
print(f"Random list with {random_count} numbers:", random_list)

# p. Will the code work when the number of random numbers is not equal to 10?
# Yes, the code will work with any number of integers.
```

```python
#Exercise 3: Working On A Paragraph
paragraph = """Lorem ipsum dolor sit amet, consectetur adipiscing elit. Integer nec odio. Praesent libero. 
Sed cursus ante dapibus diam. Sed nisi. Nulla quis sem at nibh elementum imperdiet. Duis sagittis ipsum. 
Praesent mauris. Fusce nec tellus sed augue semper porta. Mauris massa."""

# a. Number of characters
num_characters = len(paragraph)
print("Number of characters:", num_characters)

# b. Number of sentences
sentences = paragraph.split('. ')
num_sentences = len(sentences)
print("Number of sentences:", num_sentences)

# c. Number of words
words = paragraph.split()
num_words = len(words)
print("Number of words:", num_words)

# d. Number of unique words
unique_words = set(words)
num_unique_words = len(unique_words)
print("Number of unique words:", num_unique_words)

# Bonus: Number of non-whitespace characters
num_non_whitespace = len(paragraph.replace(" ", "").replace("\n", ""))
print("Number of non-whitespace characters:", num_non_whitespace)

# Bonus: Average amount of words per sentence
avg_words_per_sentence = num_words / num_sentences
print("Average words per sentence:", avg_words_per_sentence)

# Bonus: Amount of non-unique words
num_non_unique_words = num_words - num_unique_words
print("Amount of non-unique words:", num_non_unique_words)
```

```python
#Exercise 4: Frequency Of The Words
from collections import Counter

input_text = "New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3."
words = input_text.split()
word_count = Counter(words)

for word, count in word_count.items():
    print(f"{word}:{count}")
```
