```python
# Exercise 1: Convert Lists Into Dictionaries

keys = ['Ten', 'Twenty', 'Thirty']
values = [10, 20, 30]

# Using dictionary comprehension
result_dict = {keys[i]: values[i] for i in range(len(keys))}

print(result_dict)
```

```python
# Exercise 2: Cinemax

# Given family dictionary
family = {"rick": 43, 'beth': 13, 'morty': 5, 'summer': 8}

total_cost = 0

for person, age in family.items():
    if age < 3:
        ticket_price = 0
    elif 3 <= age <= 12:
        ticket_price = 10
    else:
        ticket_price = 15
    
    total_cost += ticket_price
    print(f"{person.capitalize()} has to pay ${ticket_price}.")

print(f"Total cost for the family: ${total_cost}")
```

```python
# Exercise 3: Zara

# Step 1: Information about Zara
zara_info = {
    "name": "Zara",
    "creation_date": 1975,
    "creator_name": "Amancio Ortega Gaona",
    "type_of_clothes": ["men", "women", "children", "home"],
    "international_competitors": ["Gap", "H&M", "Benetton"],
    "number_stores": 7000,
    "major_color": {
        "France": ["blue"],
        "Spain": ["red"],
        "US": ["pink", "green"]
    }
}

# Step 2: Modify information
zara_info["number_stores"] = 2
zara_info["country_creation"] = "Spain"

if "international_competitors" in zara_info:
    zara_info["international_competitors"].append("Desigual")

# Delete creation date information
del zara_info["creation_date"]

# Print statements
print(f"Zara's clients are mainly interested in {', '.join(zara_info['type_of_clothes'])} clothing.")
print(f"The last international competitor listed is: {zara_info['international_competitors'][-1]}.")
print(f"The major clothes colors in the US are: {', '.join(zara_info['major_color']['US'])}.")
print(f"The number of key-value pairs in the dictionary is: {len(zara_info)}")
print(f"The keys of the dictionary are: {', '.join(zara_info.keys())}")

# Step 3: Create more_on_zara dictionary
more_on_zara = {
    "creation_date": 1975,
    "number_stores": 10000
}

# Merge more_on_zara into zara_info
zara_info.update(more_on_zara)

print(f"The value of number_stores after update: {zara_info['number_stores']}")
```

```python
# Exercise 4: Disney Characters

users = ["Mickey", "Minnie", "Donald", "Ariel", "Pluto"]

# Recreate result 1
disney_users_A = {users[i]: i for i in range(len(users))}
print(disney_users_A)

# Recreate result 2
disney_users_B = {i: users[i] for i in range(len(users))}
print(disney_users_B)

# Recreate result 3
disney_users_C = {name: i for i, name in enumerate(sorted(users))}
print(disney_users_C)

# Recreate result 1 for names containing 'i' and starting with 'm' or 'p'
disney_users_filtered = {name: i for i, name in enumerate(users) if 'i' in name or name.startswith(('m', 'p'))}
print(disney_users_filtered)
```

```python
# Exercise 5: Random

import random

def compare_numbers(user_number):
    random_number = random.randint(1, 100)
    print(f"Random number generated: {random_number}")
    
    if user_number == random_number:
        print("Success! You guessed the correct number.")
    else:
        print(f"Fail! You guessed {user_number}, but the correct number was {random_number}.")

# Example usage:
user_number = int(input("Enter a number between 1 and 100: "))
compare_numbers(user_number)
```

```python
# Exercise 6: Temperature Advice

import random

def get_random_temp(season):
    if season == 'summer':
        return random.uniform(24, 40)  # Floating-point temperature
    elif season == 'spring' or season == 'autumn':
        return random.randint(16, 23)
    elif season == 'winter':
        return random.randint(-10, 16)

def main():
    season = input("Enter the current season (summer, autumn, winter, spring): ").lower()
    temperature = get_random_temp(season)
    
    print(f"The temperature right now is {temperature:.1f} degrees Celsius.")
    
    if temperature < 0:
        print("Brrr, that’s freezing! Wear some extra layers today.")
    elif 0 <= temperature < 16:
        print("Quite chilly! Don’t forget your coat.")
    elif 16 <= temperature < 24:
        print("Nice weather! Enjoy your day.")
    elif 24 <= temperature < 32:
        print("Warm day ahead. Stay hydrated.")
    elif 32 <= temperature <= 40:
        print("Hot day! Stay cool.")

if __name__ == "__main__":
    main()
```

```python
# Exercise 7: Star Wars Quiz

data = [
    {"question": "What is Baby Yoda's real name?", "answer": "Grogu"},
    {"question": "Where did Obi-Wan take Luke after his birth?", "answer": "Tatooine"},
    {"question": "What year did the first Star Wars movie come out?", "answer": "1977"},
    {"question": "Who built C-3PO?", "answer": "Anakin Skywalker"},
    {"question": "Anakin Skywalker grew up to be who?", "answer": "Darth Vader"},
    {"question": "What species is Chewbacca?", "answer": "Wookiee"}
]

def take_quiz():
    correct_answers = 0
    wrong_answers = []
    
    for question_data in data:
        user_answer = input(question_data["question"] + " ").strip().lower()
        if user_answer == question_data["answer"].lower():
            correct_answers += 1
        else:
            wrong_answers.append({
                "question": question_data["question"],
                "user_answer": user_answer,
                "correct_answer": question_data["answer"]
            })
    
    print(f"\nNumber of correct answers: {correct_answers}")
    print(f"Number of incorrect answers: {len(wrong_answers)}")
    
    if len(wrong_answers) > 3:
        print("\nYou answered more than 3 questions incorrectly. Please play again.")
    elif len(wrong_answers) > 0:
        print("\nIncorrect answers details:")
        for answer in wrong_answers:
            print(f"Question: {answer['question']}")
            print(f"Your answer: {answer['user_answer']}")
            print(f"Correct answer: {answer['correct_answer']}")
            print()
    else:
        print("\nCongratulations! You did great in the quiz.")

# Example usage:
take_quiz()
```
